```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from array import *
from scipy import stats


df = pd.read_csv("I:\VS Code\Prob Hackathon (Air Value)\city_day2.csv")
df = df.dropna(axis = 0)


f1 = pd.DataFrame(df, columns = ['City'])
f2 = pd.DataFrame(df, columns = ['Date'])
f3 = pd.DataFrame(df, columns = ['AQI'])


def estimate_coef(x, y):
	# number of observations/points
	n = np.size(x)

	# mean of x and y vector
	m_x = np.mean(x)
	m_y = np.mean (y)

	# calculating cross-deviation and deviation about x
	SS_xy = np.sum(y*x) - n*m_y*m_x
	SS_xx = np.sum(x*x) - n*m_x*m_x

	# calculating regression coefficients
	b_1 = SS_xy / SS_xx
	b_0 = m_y - b_1*m_x

	return (b_0, b_1)

def plot_regression_line(x, y, b):
	# plotting the actual points as scatter plot
	plt.scatter(x, y, color = "m",
			marker = "o", s = 30)

	# predicted response vector
	y_pred = b[0] + b[1]*x

	# plotting the regression line
	plt.plot(x, y_pred, color = "g")

	# putting labels
	plt.xlabel('x')
	plt.ylabel('y')

	# function to show plot
	plt.show()
 
def p(x,y,b):
  slope = b[1]
  c = b[0]
  y = slope*x + c
  return y

def main():
    # observations / data
    global x
    nameOfCity = input()
    PM = input()
    x = df.loc[df["City"] == nameOfCity, PM]
    y = df.loc[df["City"] == nameOfCity, "AQI"]
  
    # estimating coefficients
    b = estimate_coef(x, y)

    # plotting regression line
    plot_regression_line(x, y, b)

    x = float(input())
    print(p(x, y, b))
  
if __name__== "_main_":
    main()
 

size = 0
pm = x

if(pm > 0 and pm<= 30):
    x = df.loc[(df["PM2.5"] <= 30), "PM2.5"]
    size = x.shape[0]
elif(pm > 30 and pm <= 60):
    x = df.loc[(df["PM2.5"] <= 60) & (df["PM2.5"] > 30), "PM2.5"]
    size = x.shape[0]
elif(pm > 60 and pm <= 90):
    x = df.loc[(df["PM2.5"] <= 90) & (df["PM2.5"] > 60), "PM2.5"]
    size = x.shape[0]
elif(pm > 90 and pm <= 120):
    x = df.loc[(df["PM2.5"] <= 120) & (df["PM2.5"] > 90), "PM2.5"]
    size = x.shape[0]
elif(pm > 120 and pm <= 250):
    x = df.loc[(df["PM2.5"] <= 250) & (df["PM2.5"] > 120), "PM2.5"]
    size = x.shape[0]
else:
    x = df.loc[(df["PM2.5"] > 250), "PM2.5"]
    size = x.shape[0]


z = (0.33 * 0.072)/(size/df.shape[0])
print(z*100)
```
