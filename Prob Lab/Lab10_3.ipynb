```python
import numpy as np
import random
import seaborn as sns
s = []
for i in range(500):
    x = np.random.exponential(size = 30, scale = 0.35)
    s.append(sum(x)/len(x))
print(sum(s)/len(s))
```

Question - 2

```python
import numpy as np
import random
import seaborn as sns
s = []
h = [1, 2, 3, 4, 5, 6, 7, 8, 9]
px = [0.1,0.05,0.05,0.2,0.1,0.15,0.15,0.1, 0.1]
np.random.seed(90)

for i in range(2000):
    x = random.choices(h, weights=px, k=8)
    s.append(sum(x)/len(x))

sns.displot(s)
sum(s)/len(s)
```
